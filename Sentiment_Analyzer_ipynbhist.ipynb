```
import pandas as pd
import tweepy
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

tweets = api.search('Rahul Gandhi', count=1000)
tweets2 = api.search('Narender Modi', count = 1000)

data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])
data2 = pd.DataFrame(data=[tweet.text for tweet in tweets2], columns=['Tweets'])
display(data.head(10))
display(data2.head(10))

print(tweets[0].created_at)
print(tweets2[0].created_at)
```

```
import nltk
nltk.download('vader_lexicon')
```

```
sid = SentimentIntensityAnalyzer()


listy = []
listy2=[]

for index, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)

for index, row in data2.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy2.append(ss)
se = pd.Series(listy)
se2 = pd.Series(listy2)
data['polarity'] = se.values
data2['polarity'] = se2.values
display(data.head(100))
```

```
import numpy as np
dic =  data['polarity']
h = dict(dic)
df = pd.DataFrame(h)
lis = np.array(df.T['neg'])


dic2 =  data2['polarity']
h2 = dict(dic2)
df2 = pd.DataFrame(h2)
lis2 = np.array(df2.T['neg'])
print(lis2.mean())
print(lis.mean())
```
